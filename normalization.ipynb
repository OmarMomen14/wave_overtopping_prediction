```python
scaler = StandardScaler()
```

```python
df_inputs = df[df.columns[:-1]]
```

```python
X_scaled = scaler.fit_transform(df_inputs)
```

```python
X_scaled.shape
```

```python
df_scaled = pd.DataFrame(data = X_scaled, columns = df.columns[:-1])
```

```python
df_scaled['q'] = df['q']
```
